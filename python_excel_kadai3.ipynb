```python
import openpyxl
import pandas as pd
from openpyxl.styles import PatternFill

# 処理対象のファイル名のリストを格納
file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

# 結合後のデータを格納するための空のデータフレームを作成
df = pd.DataFrame()

# ファイルを一つずつ処理する
for file_name in file_list:
    # Excelファイルを読み込む
    temp_df = pd.read_excel(file_name)

    # データフレームの列名を指定
    temp_df.columns = ['売上年','商品','金額（千円）']

    # concat()でデータフレームを連結
    df = pd.concat([df, temp_df], ignore_index=False)


#商品/売上年ごとに金額を合計
grouped = df.groupby(['商品','売上年'])['金額（千円）'].sum().reset_index()

#エクセルファイルを作成する
writer = pd.ExcelWriter('売上集計表.xlsx')

#データフレームオブジェクトをExcelファイルに書き込む
grouped.to_excel(writer, sheet_name='売上集計',index=False)

writer.close()

wb = openpyxl.load_workbook('売上集計表.xlsx')
ws = wb['売上集計']

#シートのヘッダー部分を指定して色を付ける
ws['A1'].fill = PatternFill(patternType='solid' , fgColor='F2F2F2')
ws['B1'].fill = PatternFill(patternType='solid' , fgColor='F2F2F2')
ws['C1'].fill = PatternFill(patternType='solid' , fgColor='F2F2F2')

wb.save('売上集計表.xlsx')
```
