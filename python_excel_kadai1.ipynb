```python
import openpyxl
from datetime import datetime

#ワークブック作成
workbook =openpyxl.Workbook()
sheet = workbook.active

sheet['B2'] = "請求書"
conpanys = ["株式会社ABC","〒101-0022 東京都千代田区神田練塀町300",
           "TEL:03-1234-5678 FAX:03-1234-5678","担当者名:鈴木一郎 様"]

#宛名入力
i = 4
for conpany in conpanys:
  sheet.cell(row=i, column=2, value= conpany)
  i = i + 1

#請求書Noと日付
nodata = "0001"
sheet['F4'] = "No"
sheet['G4'] = nodata

today = datetime.today()
day = today.strftime('%Y%m%d')
sheet['F5'] = "日付"
sheet['G5'] = day

#表のヘッダーの書き込み
headers = ['商品名', '数量', '単価', '金額']
ii = 2
for header in headers:
  sheet.cell(row=10, column=ii, value= header)
  ii = ii + 1


#表のデータを書き込み
data1 = ['商品A', 2, 10000, 20000]
data2 = ['商品B', 1, 15000, 15000]

#dataの各行をワークシートに追加する
iii = 2
for data in data1:
  sheet.cell(row=11, column=iii, value= data)
  iii = iii + 1
iii = 2
for data in data2:
  sheet.cell(row=12, column=iii, value= data)
  iii = iii + 1

#合計と消費税
sheet['B15'] = "合計"
total = data1[3] + data2[3]
sheet['E13'] = total
sheet['E15'] = total
sheet['B16'] = "消費税"
tax = total * 0.1
sheet['E16'] = tax
sheet['B17'] = "税込合計"
sheet['E17'] = total+ tax

workbook.save(f"請求書_{today.strftime('%Y%m%d')}.xlsx")
```
